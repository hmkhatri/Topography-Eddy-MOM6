$\large \text{Topographic from stress and bottom friction}$

From Constantinou and Hogg (2019)
\begin{equation}
\underbrace{\langle \tau \rangle}_{\text{WS}} = \underbrace{\langle \overline{p_b\partial_x h_b} \rangle}_{\text{TFS}} + \underbrace{\langle \rho_mC_d|{\bf{u}}_n |u_n \rangle}_{\text{BD}}
\end{equation}

```python
# Grid create for computations
grid = Grid(ds_FW_beta, coords={'X': {'center': 'xh', 'right': 'xq'},
                        'Y': {'center': 'yh', 'right': 'yq'}}, periodic=['X']);

def topo_form(ds):
    
    dx = 4000.
    
    p_bot = 9.8 * (ds['hm'] * ds['zl']).sum('zl')
    
    dx_topo =  - grid.interp(grid.diff(ds['depth'], 'X', boundary='fill') / dx, 'X', boundary='fill')
    # -ve sign in d/dx(topo) is required because topography = H - depth.
    
    topo_form = p_bot * dx_topo
    
    return topo_form
```

```python
TF_FW_beta = topo_form(ds_FW_beta); TF_FW_beta = TF_FW_beta.mean(['xh', 'yh'])
TF_VW_beta = topo_form(ds_VW_beta); TF_VW_beta = TF_VW_beta.mean(['xh', 'yh']) 
TF_FW_f = topo_form(ds_FW_f); TF_FW_f = TF_FW_f.mean(['xh', 'yh']) 
TF_VW_f = topo_form(ds_VW_f); TF_VW_f = TF_VW_f.mean(['xh', 'yh'])

BD_FW_beta = ds_FW_beta['taux_bot'].mean(['xq', 'yh'])
BD_VW_beta = ds_VW_beta['taux_bot'].mean(['xq', 'yh'])
BD_FW_f = ds_FW_f['taux_bot'].mean(['xq', 'yh'])
BD_VW_f = ds_VW_f['taux_bot'].mean(['xq', 'yh'])

tau = ds_FW_beta['taux'].mean(['xh', 'yh'])
```

```python
grid_plot = plt.GridSpec(1, 2, wspace=0.2, hspace=0.2)
fig = plt.figure(figsize=[15,4])

ax = fig.add_subplot(grid_plot[0, 0])
ax.semilogx(int_tau_dy, TF_FW_beta/tau,'o', ms=8, alpha=0.7)
ax.semilogx(int_tau_dy, BD_FW_beta/tau,'o', ms=8, alpha=0.7)
ax.semilogx(int_tau_dy, TF_VW_beta/tau,'*', ms=8, alpha=0.7)
ax.semilogx(int_tau_dy, BD_VW_beta/tau,'*', ms=8, alpha=0.7)
ax.axhline(y=0.,ls='--',color='k',lw=1)
ax.set_xlabel('$\int \\tau dy$ (N/m)' , fontsize=12)
ax.legend(['TFS/WS (FW)', 'BD/WS (FW)', 'TFS/WS (VW)','BD/WS (VW)'])
ax.set_title('beta-plane',fontsize=12)
ax.grid()

ax = fig.add_subplot(grid_plot[0, 1])
ax.semilogx(int_tau_dy, TF_FW_f/tau,'o', ms=8, alpha=0.7)
ax.semilogx(int_tau_dy, BD_FW_f/tau,'o', ms=8, alpha=0.7)
ax.semilogx(int_tau_dy, TF_VW_f/tau,'*', ms=8, alpha=0.7)
ax.semilogx(int_tau_dy, BD_VW_f/tau,'*', ms=8, alpha=0.7)
ax.axhline(y=0.,ls='--',color='k',lw=1)
ax.set_xlabel('$\int \\tau dy$ (N/m)' , fontsize=12)
ax.legend(['TFS/WS (FW)', 'BD/WS (FW)', 'TFS/WS (VW)','BD/WS (VW)'])
ax.set_title('f-plane',fontsize=12)
ax.grid()
```
